```python
w1 = np.random.uniform(-1,1,size=(784, 100)) / np.sqrt(784*100)
w2 = np.random.uniform(-1,1,size=(100, 10)) / np.sqrt(100*10)
learning_rate = .00005
def forward_backward(x,y):
    global w1,w2
    y = np.array([0 if i != y else 1 for i in range(10)])
    x1 = x.reshape(28*28)
    x2 = np.array([sum(w1[j][i]*x1[j] for j in range(784)) for i in range(100)])
    x3 = np.maximum(x2,0)
    yhat = np.array([sum(w2[j][i]*x3[j] for j in range(100)) for i in range(10)])
    c = .5*np.linalg.norm(yhat-y)
    dcdyhat = yhat-y
    dcdw2 = np.array([[x3[j]*dcdyhat[i] for i in range(10)] for j in range(100)])
    dcdx3 = np.array([sum(w2[j][i]*dcdyhat[i] for i in range(10)) for j in range(100)])
    dcdx2 = np.array([(1 if x2[i] > 0 else 0)*dcdx3[i] for i in range(100)])
    dcdw1 = np.array([[x1[j]*dcdx2[i] for i in range(100)] for j in range(784)])
    w2 = w2-learning_rate*dcdw2
    w1 = w1-learning_rate*dcdw1
    return c
```

```python
def forward(x):
    global w1,w2
    x1 = x.reshape(28*28)
    x2 = np.array([sum(w1[j][i]*x1[j] for j in range(784)) for i in range(100)])
    x3 = np.maximum(x2,0)
    yhat = np.array([sum(w2[j][i]*x3[j] for j in range(100)) for i in range(10)])
    return np.argmax(yhat)
```

```python
loss = []

for i in range(5000):
    loss.append(forward_backward(trainx[i],trainy[i]))

plt.plot(loss)
plt.show()
```
